# SSD

##### YOLO vs SSD
- YOLO는 448x448, SSD는 300x300 크기의 이미지를 사용
- YOLO는 FCL를 사용했지만 SSD는 1x1 Convolution Layer 사용으로 FCL 제거
- YOLO는 마지막 feature map으로 Detection 수행, SSD는 마지막 featur map에서 Extra Feature Layers 수행


### 특징
- SSD(Single Shot MultiBox Detection)는 이미지 내의 객체를 감지하기 위해 이미지의 다양한 위치에서 여러 개의 경계 상자를 예측
- Multi-box Detection: 이미지에서 사전에 정의된 경계 상자를 설정하고, 이 경계 상자 안에 객체가 존재하는지를 예측하여 객체를 인식
- Extra Convolution Layers에 나온 Feature map들 모두 Detection 수행
  - 6개의 서로 다른 scale의 feature map 사용
  - 큰 feature map(early stage feature map)에서 작은 물체 탐지
  - 작은 feature map(late stage feature map)에서는 큰 물체 탐지
- Fully Connected Layer 대신 Convolution Layer 사용하여 속도 향상
- Default Box 사용(anchor box): 서로 다른 scale과 비율을 가진 미리 계산된 box 사용

### Pipeline

##### Network
- VGG-16(Backbone) + Extra Convolution Lyaers
- Input Image: 300x300

##### Multi-Scale Feature Maps
- $N_{Bbox}$: 박스의 개수
- offset은 (cx, cy, w, h)을 포함 
- $N_{c}$는 박스별 예측 클래스: num classes(20) + background(1)


##### Box 개수 정하기
- 6개의 박스
- feature map의 각 cell마다 서로 다름 scale, 비율을 가진 미리 정해진 box생성
- Faster R-CNN의 anchor Box와 유사
- Different scale per feature maps: 최솟값과 최댓값을 정해 스케일을 점점 키워나감
- Different aspect ratio: 박스의 비율을 가로, 세로 별로 지정함


##### Training
- Hard Negative Mining
- Non Maximum suppression

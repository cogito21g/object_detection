# YOLOv1

### 특징
- Region Proposal 단계가 없음
- 이미지에서 Bounding Box와 Classification을 동시에 진행
  - 이미지, 물체를 전체적으로 관찰하여 추론(맥락적 이해가 높아짐)

### Pipeline

##### Network
- GoogLeNet 변형
  - 24rodml Convolution Layer로 특징 추출
  - 2개의 Fully Connected Layer로 box의 좌표값 및 확률 계산



##### 
- 이미지를 SxS Grid로 나누어 각 Grid 영역마다 B=2개의 Bounding Box와 Confidence Score 계산
  - Confidence Score = $Pr(Object) x IOU^{truth}_{pred}$
  
- 각 Gird별 C=20개의 Class에 대한 확률을 계산
  - conditional class probability = $Pr(Class_{i}|Object)$
  


##### 30 Channel 구성요소



##### inference 과정
1) Gird마다 각각의 Bounding Box 수행 -> 49개의 Grid x 2개의 BBox로 98개의 BBox가 나옴
2) 각 BBox의 class score를 확인하여 threshold보다 작은 값은 0으로 변경 
3) class score를 descending한 뒤 NMS를 거쳐 남은 박스를 그려줌

### 장점

- Faster R-CNN에 비해 6배 빠른 속도
- 다른 real-time detector에 비해 2배 높은 정확도
- 이미지 전체를 보기 때문에 클래스와 사진에 대한 맥락적 정보를 가지고 있음
- 물체의 일반화된 표현을 학습
- 사용된 dataset외 새로운 도메인에 대한 이미지에 대한 좋은 성능을 보임

### 단점

- 7x7 그리드 영역으로 나눠 Bounding box prediction 진행  → 그리드보다 작은 크기의 물체 검출 불가능
- 신경망을 통과하며 마지막 feature만 사용  → 정확도 하락
